```python
import pandas as pd
```

```python
college = pd.read_csv('/content/AA2.csv')
college.head()
```

```python
college.describe()
```

```python
print("Columns name: \n", college.columns)
```

```python
college.isnull()
```

```python
college = college.fillna(college.mean(numeric_only = True))
```

```python
for col in college.select_dtypes(include=['object', 'category']).columns:
    college[col] = college[col].fillna("Unknown")
```

```python
college['weight'] = college['weight'].astype(str)
```

```python
college['weight_cleaned'] = college['weight'].str.extract(r'(\d+)')
```

```python
college['weight_cleaned'] = college['weight_cleaned'].astype(float)
```

```python
college['weight_cleaned'] = pd.to_numeric(college['weight_cleaned'], errors='coerce')
college['weight_cleaned'] = college['weight_cleaned'].fillna(college['weight_cleaned'].mean())
```

```python
college.isnull().sum()
```

```python
college['Total_Calories'] = (college['calories_chicken'] + college['calories_scone'] + college['tortilla_calories'] + college['turkey_calories'] + college['waffle_calories'])
```

```python
print(college[['weight', 'weight_cleaned', 'Total_Calories']].head())
```

```python
num_cols = ['GPA', 'calories_chicken', 'calories_scone', 'tortilla_calories', 'turkey_calories', 'waffle_calories', 'weight_cleaned']

for col in num_cols:
    college[col] = pd.to_numeric(college[col], errors='coerce')
```

```python
college['Gender_cleaned'] = college['Gender'].map({1: 'Male', 2: 'Female'})
```

```python
college[['GPA', 'weight_cleaned', 'Gender', 'Gender_cleaned']].head()
```

```python
text_cols = college.select_dtypes(include=['object']).columns
```

```python
for col in text_cols:
  college[col] = college[col].str.title()
```

```python
print(college[['comfort_food', 'fav_cuisine', 'type_sports']].head())
```

```python
import matplotlib.pyplot as plt
```

```python
replace_map = {
    # Italian group
    'I Really Love Italian Food': 'Italian',
    'Italian Or Chinese': 'Italian',
    'Italian/American': 'Italian',
    'Italian And Chinese': 'Italian',
    # American group
    'Macaroni And Cheese': 'American',
    'Sub Sandwiches': 'American',
    'Anything American': 'American',
    'American-Chinese': 'American',
    # Chinese group
    'Orange Chicken And Chow Main Noodles': 'Chinese',
    'Authentic Chinese And Vietnamese Food': 'Chinese',
    'Chinese Cuisine (General Asian)': 'Chinese',
    # Indian group
    'Indian Food - Samosas And Rice': 'Indian',
    # Middle Eastern group
    'Lebanese And Greek': 'Middle Eastern',
    'Arabic Cuisine': 'Middle Eastern',
    # Unknown / unclear
    'I Don’T Have One': 'Unknown',
    'I Do Not Like Cuisine': 'Unknown',
    'And The Usual': 'Unknown',
    'Not Sure': 'Unknown',
    'Unknown.': 'Unknown',
    # Asian grouping
    'Vietnamese Food': 'Asian',
    'Japanese': 'Asian',
    'Thai/Chinese': 'Asian',
    # Composite
    'Mexican And Chinese Food': 'Mexican/Chinese',
    'Any Type Of Colombian Food': 'Colombian',
}
```

```python
college['fav_cuisine'] = college['fav_cuisine'].replace(replace_map)
```

```python
cuisine_counts = college['fav_cuisine'].value_counts().sort_values(ascending=False)
```

```python
college['fav_cuisine'] = college['fav_cuisine'].astype(str).str.strip().str.title()
```

```python
plt.figure(figsize=(10,6))
cuisine_counts.plot(kind='bar', color='orchid')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Favorite Cuisine (Grouped & Cleaned)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
college['Total Calories'] = (college['calories_chicken'] + college['calories_scone'] + college['tortilla_calories'] + college['turkey_calories'] + college['waffle_calories'])
```

```python
college['GPA'] = pd.to_numeric(college['GPA'], errors = 'coerce')
```

```python
college_line = college[['GPA', 'Total_Calories']].dropna()
```

```python
college_line = college_line.sort_values(by = 'Total_Calories')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(college_line['Total_Calories'], college_line['GPA'], marker='o', linestyle='-', color='skyblue')
plt.title('GPA vs. Daily Caloric Intake')
plt.xlabel('Total Daily Caloric Intake (kcal)')
plt.ylabel('GPA')
plt.grid(True)
plt.tight_layout()
plt.show()
```
